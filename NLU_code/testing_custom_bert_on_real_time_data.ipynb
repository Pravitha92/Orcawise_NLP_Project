```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
import torch
import random
import torchmetrics
import numpy as np
import pandas as pd
from transformers import BertTokenizer
from torch.utils.data import DataLoader, TensorDataset
from transformers import BertForSequenceClassification
from nlu_pretrained_model import OpenIEExtractor
from nlu_custom_model_old import CustomBertModel1
from nlu_custom_model import CustomBertModel
from nlu_custom_model_final import CustomBertModel_final
```

```python
df = pd.read_csv('data/book2.csv', encoding='latin-1')

predictions = []
predictions1 = []
predictions2 = []
predictions3 = []
accuracy_values = []
accuracy_values1 = []
accuracy_values2 = []
accuracy_values3 = []
```

```python
obj = CustomBertModel1(model_name="bert-base-uncased", num_labels=4, checkpoint_path='version_0/checkpoints/epoch=1-step=356.ckpt')

obj1 = CustomBertModel(model_name="bert-base-uncased", num_labels=4, checkpoint_path='version_0/checkpoints/best_model.pth')

obj2 = CustomBertModel_final(model_name="bert-base-uncased", num_labels=4, checkpoint_path='version_0/checkpoints/best_model_final.pth')

obj3 = OpenIEExtractor()
```

```python
# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    # Get the sentence from the 'sentence' column
    sentence = row['sentence']

    # Predict the relation for the current sentence
    predicted_relation = obj.predict_relation(sentence)
    predicted_relation1 = obj1.predict_relation(sentence)
    predicted_relation2 = obj2.predict_relation(sentence)
    predicted_relation3 = obj3.extract_relations(sentence)
    

    # Append the predicted relation to the list
    predictions.append(predicted_relation)
    predictions1.append(predicted_relation1)
    predictions2.append(predicted_relation2)
    predictions3.append(predicted_relation3)

    # Compare the predicted relation with the 'ground_truth' column
    ground_truth = row['ground_truth']
    accuracy = 1 if predicted_relation == ground_truth else 0
    accuracy1 = 1 if predicted_relation1 == ground_truth else 0
    accuracy2 = 1 if predicted_relation2 == ground_truth else 0

    # Append the accuracy value to the list
    accuracy_values.append(accuracy)
    accuracy_values1.append(accuracy1)
    accuracy_values2.append(accuracy2)
```

```python
df['custom_bert'] = predictions
df['acc_custom'] = accuracy_values

df['improve_bert1'] = predictions1
df['acc_custom_improve1'] = accuracy_values1

df['improve_bert2'] = predictions2
df['acc_custom_improve2'] = accuracy_values2

df['pretrained'] = predictions3

df
```

```python
def map_relation(value):
    if value is None:
        return 'noRelation'
    elif any(keyword in value for keyword in ['employs', 'employed', 'player', 'serves','crafted','assists customers','contributing']):
        return 'employedBy'
    elif any(keyword in value for keyword in ['manager', 'managed', 'directs', 'manages']):
        return 'managerOf'
    elif any(keyword in value for keyword in ['located', 'is in', 'situated', 'location', 'headquarters of']):
        return 'locatedAt'
    else:
        return 'noRelation'
        
df['acc_pre'] = df['pretrained'].apply(map_relation)
df
```

```python
# Define the categories for 'acc_pre'
categories = ['employedBy', 'managerOf', 'locatedAt', 'noRelation']

# Convert 'acc_pre' column to categorical type
df['acc_pre'] = pd.Categorical(df['acc_pre'], categories=categories)

# Create a new column 'acc_pre_binary' based on the comparison with 'ground_truth'
df['acc_pre_binary'] = np.where(df['ground_truth'] == df['acc_pre'], 1, 0)
df
```

```python
# Compute the accuracy percentage
accuracy_percentage1 = df['acc_custom'].mean() * 100
accuracy_percentage2 = df['acc_custom_improve1'].mean() * 100
accuracy_percentage3 = df['acc_custom_improve2'].mean() * 100
accuracy_percentage4 = df['acc_pre_binary'].mean() * 100
```

```python
s_accuracy_percentages1 = df.groupby('ground_truth')['acc_custom'].mean() * 100
s_accuracy_percentages2 = df.groupby('ground_truth')['acc_custom_improve1'].mean() * 100
s_accuracy_percentages3 = df.groupby('ground_truth')['acc_custom_improve2'].mean() * 100
s_accuracy_percentages4 = df.groupby('ground_truth')['acc_pre_binary'].mean() * 100

print(f"Total Accuracy of Custom Bert: {accuracy_percentage1:.2f}%")
print("Specific Accuracy of Custom Bert:")
print(s_accuracy_percentages1)

print("#############################################################################")
print(f"Total Accuracy of Custom Bert Improve with original data: {accuracy_percentage2:.2f}%")
print("Specific Accuracy of Custom Bert Improve1:")
print(s_accuracy_percentages2)

print("#############################################################################")
print(f"Total Accuracy of Custom Bert Improve with original data: {accuracy_percentage3:.2f}%")
print("Specific Accuracy of Custom Bert Improve2:")
print(s_accuracy_percentages3)

print("#############################################################################")
print(f"Total Accuracy of pre-trained model: {accuracy_percentage4:.2f}%")
print("Specific Accuracy of Custom Bert Improve:")
print(s_accuracy_percentages4)
```
